#How to graph a function f(x)
This time we'll se how to graph a function by defining its domain, the function itself and then using the libraties to plot it

```
# Let's import libraties
import numpy as np
import matplotlib.pyplot as plt

# First we define de domain (x0,xf,n)
x=np.linspace(0,5,120)

# Now we define the function
k=1.
f=x*np.sin(k*x)
g=np.cos(k*x)

# Now let's make the graph
plt.plot(x,f,color='r')
plt.plot(x,g,color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x)=x sin(kx)')
plt.grid()
plt.show()
```

#Calculations of numerical derivatives
We're going to calculate the fist derivative of the function
 $$
 f(x)=sin(kx)
 $$
  by using the finite-differences method

```
#Fist we'll define the physical doman

x0=0 # Initial point of the domain (mts) 
xf=10.0 # Final point if the domain (mts)
n=200 # Number of points per wavelength
dx=(xf-x0)/(n-1) # Grid increment dx (mts)

#The physical domain turns out
x=np.linspace(x0,xf,n)
```

```
#Now we'll difine the funcion f(x)
l=7
k=2*np.pi/l
f=np.sin(k*x)
```

```
# In order to verify, we plot the function f(x)
plt.plot(x,f)
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x)=sin(kx)")
plt.grid()
plt.show()
```

Now, we'll calculate the numerical derivative
$$
f'(x)\approx \frac{f(x+dx)-d(x-dx)}{2dx}
$$
which must match with the analitical one
$$
f'(x)=k cos(kx)
$$

```
# We define the numerical and analytical derivatives in order to after compare them.
# To do this we use the comand np.zeros() which is basically to define an array full of zeros
nder=np.zeros(n) # Numerical derivative
ader=np.zeros(n) # Analitical derivative

# Time to calculate
for i in range(1,n-1):
  nder[i]=(f[i+1]-f[i-1])/(2*dx)

# The analitical one
ader=k*np.cos(k*x)

# excluding the boundaries
nder[0]=0.
nder[n-1]=0.

#Now it’s time to calculate the error (rms)
rms=np.sqrt(np.mean(nder-ader)**2)
```

```
# Now we ploting
plt.figure(figsize=(10,6)) 
plt.title('Numerical & analytical derivatives')
plt.plot(x,nder,label='Numerical derivative',marker='+',color='r')
plt.plot(x,ader,label='Analytical derivative',color='g')
plt.plot(x,nder-ader)
plt.xlabel=('x[m]')
plt.ylabel=('y[m]')
plt.grid()
plt.show()
```

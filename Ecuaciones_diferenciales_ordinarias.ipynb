# Solucion de una ED de orden 1 usando la funcion odeint

```
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def f(y,t):
  return np.e**(-t**2)

y0=0
t=np.linspace(0,1,100)
sol=integrate.odeint(f,y0,t)

# grafica la solución aproximada mediante el símbolo '+' en color rojo
plt.plot(t, sol,'--',color='r')

#grafica la solución exacta mediante línea continua en color azúl
yt=(1/3)*t**(3)
plt.plot(t,yt,color='b')
plt.xlabel('t')
plt.ylabel('x')
plt.title("x(t)")
plt.grid()
plt.show()
```

```
#Let's import libreries
import numpy as np                 #For functions like e^^x,sin(x),arrays and so on
import matplotlib.pyplot as plt    #For making graphs
from scipy.integrate import odeint #To solve de DE

# Asignation parameters values
#m=5.0
#b=0.4
#k=1.0

#Let's difine the function
def f(A,t):
  x,v=A                       # ( x , dx_dt )
  return [v,-4e14/x**2] # ( v , dv_dt )

#Setting initial conditions
f0=[2*6.371e6,0.0]

#Setting the interval and precision
t=np.linspace(0,2000,1000)

#Solving the equation and printing data
sol=odeint(f,f0,t)
#sol

#Ploting 
fig, axes = plt.subplots()
axes.plot(t, sol[:,0], 'g')
#axes.plot(t, sol[:,1], 'b')
plt.grid()
plt.show()
```

# Modelo lodka volterra resuelto por el metodo de Euler

```
import matplotlib.pyplot as plt
# Defining parameters
a=0.7 ; b=0.5 ; c=0.3 ; d=0.2
dt=0.001 ; tmax=100

# Initial conditions
t=0. ; x=1. ; y=0.5

# empty lists in which to store time and populations
t_list = [t]; x_list = [x]; y_list = [y]

# initialize lists
#t_list.append(t); x_list.append(x); y_list.append(y)

while t<tmax:
  # Iterations
  t=t+dt 
  x=x+(a*x+0.01*x**2-b*x*y)*dt 
  y=y+(-c*y+d*x*y)*dt

  # store new values in lists
  t_list.append(t)
  x_list.append(x)
  y_list.append(y)

# Plot results
p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2)
```

# Oscilador armonico amortiguado forzado

```
# Importing libreries
import numpy as np
import matplotlib.pyplot as plt

#parameters 
m=1 ; k=0.75 ; b=0.05 ; f0=1

#initial conditions
t=0 ; x=0.5 ; v=0

# Defining initial lists
x_list=[x] ; t_list=[t] ; v_list=[v]

# Step $ number of points
dt=0.01 ; n=25000

# Defining the fuction
def f(t,x,v): # df_dt
  return f0*np.sin(t)-(b/m)*v-(k/m)*x #Acceleration function

while t<n*dt:
  t=t+dt
  v=v+f(t,x,v)*dt
  x=x+v*dt

  t_list.append(t)
  v_list.append(v)
  x_list.append(x)
  
#Plot results
p=plt.plot(t_list, x_list, 'r')#, t_list, v_list, 'g', linewidth = 2)
```

# Finite Difference Method
To start we will solve the function

$$
\frac{d^2u}{dt^2}+\omega^2 u=0
$$

Using the finite differences method, this is expressing derivatives like:

$$
\frac{d^2u}{dt^2}=\frac{u(t+dt)-2u(t)-u(t-dt)}{dt^2}
$$

In code notation is:

$$
\frac{d^2u}{dt^2}=\frac{u^{n+1}-2u^n-u^{n-1}}{dt^2}
$$

So that the solution can be obtaied iterating over the older term

$$
u^{n+1}=(2-\omega ^2 dt^2)u^n+u^{n-1}
$$

```
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```
w=1. #Angular frequency
A=1. #Amplitude
n=168 #Number of points in the whole domain
T=35  #Simulation time's period

t=np.linspace(0,T,n+1)
dt=t[1]-t[0]
u=np.zeros(n+1)
er=np.zeros(n+1)

u[0]=A 
u[1]=u[0]-0.5*dt**2*w**2*u[0]

for n in range(1,n):
  u[n+1]=2*u[n]-u[n-1]-dt**2*w**2*u[n]
  er[n]=u[n]-np.cos(t[n])
plt.figure(figsize=(12, 6))
plt.plot(t,u,'.',label='Numerical')
plt.plot(t,np.cos(t),label='Analytical')
plt.plot(t,er,'--',label='Error')
#plt.plot(t,er)
plt.grid()
plt.legend()
plt.title('Cos(t)')
plt.show()

```

From the previous figure we can see a slight error in the numerical frequency which increases over time, this is more and more evident at longer time periods.\\

To study this error let us remember the solution of the harmonic oscillator
$$
\frac{d^2 u}{dt^2}+w^2 u=0 \rightarrow u=e^{iw'dt}
$$
From here we have that
$$
u^n=e^{{iw'ndt}}=e^{iw't_n}
$$
Substituting this into the differential equation in terms of finite difference method
$$
\frac{d^2u}{dt^2}=\frac{u^{n+1}-2u^{n}+u^{n-1}}{dt^2}
$$

$$
=A\frac{e^{iw'(t_n+dt)}-2e^{iw't_n}+e^{iw'(t_n-dt)}}{dt^2}
$$

$$
=Ae^{iw't_n}\frac{e^{iw'dt}+e^{-iw'dt}-2}{dt^2}
$$

$$
=2Ae^{iw't_n}\frac{cosh(iw'dt)-1}{dt^2}
$$

$$
=2Ae^{iw't_n}\frac{cos(w'dt)-1}{dt^2}
$$

Taking advantage of  $cos(x)-1=-2sin^2(x/2)$ the one above turns out

$$
=-A\frac{4}{dt^2}sin^2\left(\frac{w'dt}{2} \right)
$$

Substituting this in the DE

$$
-A\frac{4}{dt^2}sin^2\left(\frac{w'dt}{2} \right)+Aw^2e^{iw't_n}=0
$$

Which results:

$$
w'=\frac{2}{dt}sin^{-1}\left(\frac{wdt}{2} \right)
$$

Expanding in Taylor series this turns out

$$
w'=w\left( 1+\frac{1}{24}w^2dt^2\right)+O(dt^4)
$$

Here we can see that $w'>w$ due to the term $w^3dt^2/24$ and this is why the numerical frequency and the analytical does not match at long period of time. If we want to make them be the same we need to reduce the term $wdt$ and this is gotten by meshing finer.
In addition, we can express this in terms of the number of points taking into account that $P=N_{p}dt$ with P the period and $N_p$ the number of points per period, at the same time $P=2\pi/w$, then using this we can express the term $wdt$ as
$$
wdt=\frac{2\pi}{N_p}
$$

Now we will do a subroutine that solve the diferential equation and depends opon the parameters A, w, dt, T.

```
def solver(A,w,dt,T):
  # parameters
  dt = float(dt)
  Nt = int(round(T/dt))
  u = np.zeros(Nt+1)
  t = np.linspace(0, Nt*dt, Nt+1)
  # Initial conditions
  u[0] = A
  u[1] = u[0] - 0.5*dt**2*w**2*u[0]
  # for loop
  for n in range(1, Nt):
    u[n+1] = 2*u[n] - u[n-1] - dt**2*w**2*u[n]
  return u, t
```

Defining the analytical solution:


```
def u_exact(t, A, w):
  return A*np.cos(w*t)
```

Now we compare both solutions graphically

```
def visualize(u,t,A,w):
  plt.plot(t,u,'r--o')
  t_fine=np.linspace(0,t[-1],1001)
  u_e=u_exact(t_fine,A,w)
  plt.plot(t_fine, u_e, 'b-')
  plt.legend(['numerical', 'exact'], loc='upper left')
  plt.xlabel('t')
  plt.ylabel('u')
  dt = t[1] - t[0]
  plt.title('dt=%g' % dt)
  umin = 1.7*u.min(); umax = -umin
  plt.axis([t[0], t[-1], umin, umax])
  #plt.savefig('tmp1.png'); plt.savefig('tmp1.pdf')
```

```
def convergence_rates(m, solver_function, num_periods=8):
  """
  Return m-1 empirical estimates of the convergence rate
  based on m simulations, where the time step is halved
  for each simulation.
  solver_function(I, w, dt, T) solves each problem, where T
  is based on simulation for num_periods periods.
  """
  from math import pi
  w = 0.35; I = 0.3
  # just chosen values
  P = 2*pi/w
  # period
  dt = P/30
  # 30 time step per period 2*pi/w
  T = P*num_periods
  dt_values = []
  E_values = []
  for i in range(m):
    u, t = solver_function(I, w, dt, T)
    u_e = u_exact(t, I, w)
    E = np.sqrt(dt*np.sum((u_e-u)**2))
    dt_values.append(dt)
    E_values.append(E)
    dt = dt/2
  
  r = [np.log(E_values[i-1]/E_values[i])/
      np.log(dt_values[i-1]/dt_values[i])
      for i in range(1, m, 1)]
  return r, E_values, dt_values
  len(E)
```

# Solution of the wave equation without source term
$$
\frac{\partial ^2 u}{\partial t^2}=c^2\frac{\partial ^2 u}{\partial x^2}
$$

Where the initial y boundary conditions are

$$
u(0,t)=0\ ,\ \ \ u(L,t)=0\ ,\ \ \ u(x,0)=f(x)\ , \ \ \ u_{t}(x,0)=0
$$

And the explicit values are
$$
L=1\ meter, \ \ \ c=0.5\ m/s, \ \ \ T=1\ seg.
$$

```
import numpy as np
import matplotlib.pyplot as plt

# Parameters
nx=100
nt=nx
L=1.
T=1.
x=np.linspace(0,L,nx)
t=np.linspace(0,T,nt)
c=0.5

dx=L/(nx-1)
dt=T/(nt-1)
C=c*dt/dx

Nt=len(t)-1
C2=c**2
```

We will simulate a string stretched from the center and fixed from the ends

```
# Defining the initial wave profile u(x,0)=f(x)
a=0.3
x1 = np.linspace(0,0.5)
x2 = np.linspace(0.5,1)
y1 = a*x1
y2 = a*(1-x2)
x3 = np.concatenate((x1,x2))
y3 = np.concatenate((y1,y2))

plt.plot(x3,y3)
plt.grid()
plt.ylim((0,1))
plt.show()

```

We will use  the following representations:
$$
u_{i}^{n+1}\ \ \ \ \rightarrow\ \ \ \ \ u_{new}[i]
$$
$$
u_{i}^{n}\ \ \ \ \ \rightarrow \ \ \ \  u[i]
$$
$$
u_{i}^{n-1}\ \ \ \ \rightarrow\ \ \ \ \ u_{old}[i]   
$$

To compute the vectors in out simulation

```
# Let us take u0 as a copy of the vector that represents to f(x)
u=y3.copy()
unew=np.zeros(len(u))
uold=np.zeros(len(u))

# Let's store u_new at t=0.
for i in range(1, nx-1):
  unew[i] = u[i] - 0.5*C2*(u[i+1] - 2*u[i] + u[i-1])
# We already have the vectors u and u_new at the begining of our simulation,
# This is u^0=(u00,u10,....u0nx) & u^1=(u01,u11,...,unx1).
# Now we must iterate over time to obtain all vectors at t=1,2,3,....T
```

Now we need that the new vector, pass to be the actual and the actual becomes the old one.
$$
u_{new} \ \ \ \ \ \rightarrow \ \ \ \ \ u
$$
$$
u \ \ \ \ \ \rightarrow \ \ \ \ \ u_{old}
$$

```
# --- Option 1 --- Vectorizing
#uold[:] = u
#u[:]=unew

# --- Option 2 --- Using loops
for i in range(1,nx-1):
  uold[i]=u[i]
  u[i]=unew[i]

for n in range(1, nt-1):
# Update all inner mesh points at time t[n+1]
  for i in range(1, nx-1):
    unew[i] = 2*u[i] - uold[i] + C2*(u[i+1] - 2*u[i] + u[i-1])
  # Insert boundary conditions
  #u[0] = 0; u[nx] = 0
  # Switch variables before next step
  uold[:], u[:] = u, unew
```
